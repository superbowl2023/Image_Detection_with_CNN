```python
# importing the libraries
import tensorflow as tf
print(f"Tensorflow Version: {tf.__version__}")
```

Part 1 - Data Preprocessing

```python
# Preprocessing the training set
ImageDataGenerator = tf.keras.preprocessing.image.ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
training_set = train_datagen.flow_from_directory(
    'dataset/training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
```

```python
test_datagen = ImageDataGenerator(
    rescale=1./255,
)
test_set = test_datagen.flow_from_directory(
    'dataset/test_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
```

Part 2 Building the CNN

```python
#Initializing the CNN
cnn = tf.keras.models.Sequential()
```

```python
## 1. Convolution
cnn.add(tf.keras.layers.Conv2D(input_shape=[64, 64, 3], filters=32, kernel_size=3, activation='relu'))
```

```python
## 2. Pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
# Adding a second convolutional layer
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
## 3. Flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
## 4. Full Connection
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
## 5. Output Layer. We use just one unit for output. Sigmoid for binary classification (for multiclass use softmax )
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Part 3 - Training the CNN


```python
# Compiling the CNN
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Training the CNN on the Training set and evaluating it on the test set.
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

```python
import numpy as np
image = tf.keras.preprocessing.image

test_image = image.load_img('single_prediction/cat_or_dog_2.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```
